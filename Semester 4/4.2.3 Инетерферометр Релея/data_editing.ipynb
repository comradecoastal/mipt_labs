```python
import pandas as pd
import numpy as np
```

```python
with open("gas.csv") as f:
    sheet = pd.read_csv(f)
    
sheet.insert(0, "t", sheet["min"] * 60 + sheet["sec"])
sheet.drop(columns=["min","sec"], inplace=True)

with open("gas_t.csv","w") as f:
    sheet.to_csv(f, index=False)
```
