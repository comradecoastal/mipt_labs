```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import norm


from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['mathtext.fontset'] = 'dejavuserif' 
rcParams['figure.figsize'] = (10, 7)
rcParams['figure.dpi'] = 300
rcParams['axes.grid'] = True
rcParams['font.size'] = 14
# rcParams
```

```python
r = np.array([27 , 39  , 47  , 54  , 60  , 67  , 71  , 76  , 81  , 84  , 88])
r2 = r**2
n = np.arange(1, 12)

2 * r
```

```python
# plt.scatter(n, r2)
plt.errorbar(n, r2, 2 * r, np.zeros(11), 'none', capsize=3)

arg, cov = np.polyfit(n, r2, 1, w=1/r, cov=True)
err = np.sqrt(cov.diagonal())
print(arg, err)
f = np.poly1d(arg)
plt.plot(n, f(n), ls='--', label="Аппроксимация зависимости $r^2(n)$\n$y = ({:2.0f} \pm {:2.0f})x$".format(arg[0], err[0]))

plt.xlabel("Порядок минимума, $m$")
plt.ylabel("Квадрат радиуса, $r^2$ мм$^2$")

plt.legend()

plt.savefig("plot.pdf")
```
