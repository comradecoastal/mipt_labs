```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import norm


from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['mathtext.fontset'] = 'dejavuserif' 
rcParams['figure.figsize'] = (10, 7)
rcParams['figure.dpi'] = 300
rcParams['axes.grid'] = True
rcParams['font.size'] = 14
# rcParams
```

```python
with open("vac.csv") as f:
    vac = pd.read_csv(f)
with open("afc.csv") as f:
    afc = pd.read_csv(f)

vac.adc = np.abs(vac.adc)
```

```python
vac
```

```python
afc
```

```python
plt.plot(vac.dac * 1e-3, vac.adc * 1e-3)

plt.xlabel("Показания ЦАП, мВ")
plt.ylabel("Показания АЦП, мВ")

plt.title(r"ВАХ молекулярно-электронного преобразователя")

plt.savefig("vac.pdf")
```

```python
plt.plot(afc.hz, afc.dac, label="Размах ЦАП")
plt.plot(afc.hz, afc.adc, label="Размах АЦП")
plt.scatter(afc.hz, afc.dac, marker='.')
plt.scatter(afc.hz, afc.adc, marker='.')

plt.title(r"АЧХ молекулярно-электронного преобразователя")
plt.xlabel(r"Частота колебаний, Гц")
plt.ylabel(r"Сила тока, мкВ")

plt.legend()
plt.savefig("afc.pdf")
```

```python
plt.plot(afc.hz, afc.dac, label="Размах ЦАП")
plt.plot(afc.hz, afc.adc, label="Размах АЦП")
plt.scatter(afc.hz, afc.dac, marker='.')
plt.scatter(afc.hz, afc.adc, marker='.')

plt.title(r"Логарифм АЧХ молекулярно-электронного преобразователя")
plt.xlabel(r"Частота колебаний, Гц")
plt.ylabel(r"Логарифм силы тока, $\ln$мкВ")

plt.yscale("log")
plt.ylim((1e3, 1e5))

plt.legend()
plt.savefig("afc_log.pdf")
```
