```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import norm


from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['mathtext.fontset'] = 'dejavuserif' 
rcParams['figure.figsize'] = (10, 7)
rcParams['figure.dpi'] = 300
rcParams['axes.grid'] = True
rcParams['font.size'] = 14
# rcParams
```

```python
with open("vac.csv") as f:
    vac = pd.read_csv(f)
with open("afc.csv") as f:
    afc = pd.read_csv(f)

vac.adc = np.abs(vac.adc)
```

```python
plt.plot(vac.dac * 1e-3, vac.adc * 1e-3)

plt.xlabel(" , ")
plt.ylabel(" , ")

plt.title(r" - ")

plt.savefig("vac.pdf")
```

```python
plt.plot(afc.hz, afc.dac, label=" ")
plt.plot(afc.hz, afc.adc, label=" ")
plt.scatter(afc.hz, afc.dac, marker='.')
plt.scatter(afc.hz, afc.adc, marker='.')

plt.title(r" - ")
plt.xlabel(r" , ")
plt.ylabel(r" , ")

plt.legend()
plt.savefig("afc.pdf")
```

```python
plt.plot(afc.hz, afc.dac, label=" ")
plt.plot(afc.hz, afc.adc, label=" ")
plt.scatter(afc.hz, afc.dac, marker='.')
plt.scatter(afc.hz, afc.adc, marker='.')

plt.title(r"  - ")
plt.xlabel(r" , ")
plt.ylabel(r"  , $\ln$")

plt.yscale("log")
plt.ylim((1e3, 1e5))

plt.legend()
plt.savefig("afc_log.pdf")
```
