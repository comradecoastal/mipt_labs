```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import norm


from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['mathtext.fontset'] = 'dejavuserif' 
rcParams['figure.figsize'] = (10, 7)
rcParams['figure.dpi'] = 300
rcParams['axes.grid'] = True
rcParams['font.size'] = 14
# rcParams
```

```python
u_mkp = np.array([0.32, 0.60, 0.90, 1.20, 1.50, 1.80, 2.00])
i_mkp = np.array([1.33, 2.53, 3.70, 5.00, 6.34, 7.66, 8.80])

plt.scatter(u_mkp, i_mkp)
plt.ylabel(r"$I_\mathcal{МКП}$, мА")
plt.xlabel(r"$U_\mathcal{МКП}$, В")
arg = np.polyfit(u_mkp, i_mkp, 1)
f = np.poly1d(arg)
plt.plot(u_mkp, f(u_mkp), ls='--', lw=1, label="Линейная аппроксимация зависимости\n$y = {:.2f} \\cdot x {:.2f}$".format(arg[0], arg[1]))
plt.legend()

plt.savefig("plot2.pdf")
```
