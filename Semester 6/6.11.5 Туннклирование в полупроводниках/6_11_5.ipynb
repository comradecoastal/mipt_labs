```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['mathtext.fontset'] = 'dejavuserif' 
rcParams['figure.figsize'] = (10, 7)
rcParams['figure.dpi'] = 100
rcParams['axes.grid'] = True
rcParams['font.size'] = 14
# rcParams
```

```python
with open("data.csv") as f:
    data = pd.read_csv(f)
```

```python
np.max(data.I)
```

```python
x, y = np.array(data.U), np.array(data.I)
I_p, n_p = np.max(y[:10]), np.argmax(y[:10]) + 1
I_v, n_v = np.min(y[10:]), np.argmin(y[10:]) + 10
U_p, U_v = x[n_p], x[n_v]
U_f = (x[-2] * (y[-2] - I_p) + x[-3] * (I_p - y[-3])) / (y[-2] - y[-3])
print(I_p, U_p, I_v, U_v, U_f)

plt.scatter(x, y, color="black")

plt.yticks([0, 1, 2, 3, 5] + [I_v, I_p], [0, 1, 2, 3, 5] + ["$I_v$", "$I_p$"])
plt.xticks([0, 100, 200, 300] + [U_v, U_p, U_f], [0, 100, 200, 300] + ["$U_v$", "$U_p$", "$U_f$"])
plt.xlabel("Напряжение на диоде $U$, мВ")
plt.ylabel("Ток текущий через диод $I$, мА")

plt.savefig("plot.pdf")
```
