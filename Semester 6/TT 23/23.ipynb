```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['mathtext.fontset'] = 'dejavuserif' 
rcParams['figure.figsize'] = (10, 7)
rcParams['figure.dpi'] = 100
rcParams['axes.grid'] = True
rcParams['font.size'] = 14
# rcParams
```

```python
with open("data.csv") as f:
    data = pd.read_csv(f)
    
data["l"] = data.N * 0.5 # mm

data
```

```python
x, y = -data.l, data.U
# plt.scatter(x, y)
plt.errorbar(x, y, 0.1, 0, "none", capsize=2)
plt.xlabel("Изменение расстояния $\Delta l$, мм")
plt.ylabel("Напряжение на участве образца $U_x$, мВ")

arg, cov = np.polyfit(x, y, 1, cov=True)
err = np.sqrt(np.diag(cov))
print(arg, err)

plt.plot(x, arg[0] * x + arg[1], ls="--", label=f"$y = ({arg[0]:.2f} \pm {err[0]:.2f}) x + {arg[1]:.1f} \pm {err[1]:.1f}$")

plt.legend()
plt.savefig("plot.pdf")
```
