```python
import numpy as np
import pandas as pd
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['mathtext.fontset'] = 'dejavuserif' 
import matplotlib.pyplot as plt
```

```python
# функция мнк

def lsf(x, y, functional=True):
    x_avg = np.average(x)
    y_avg = np.average(y)
    xy_avg = np.average(x * y)
    x2_avg = np.average(np.square(x))
    y2_avg = np.average(np.square(y))
    count = len(x)
    
    print('x=', x_avg, 'y=', y_avg)
    print('x2=', x2_avg, 'y2=', y2_avg, 'xy=', xy_avg)
    
    b = (xy_avg - x_avg * y_avg) / (x2_avg - x_avg ** 2)
    b_error = ((y2_avg - y_avg ** 2)/(x2_avg - x_avg ** 2) - b ** 2) ** 0.5 / count ** 0.5
    a = y_avg - b * x_avg
    a_error = b_error * (x2_avg - x_avg ** 2) ** 0.5
    
    if functional:
        f = lambda x: b * x + a
        return f, (b, a, b_error, a_error)
    
    return b, a, b_error, a_error
```

```python
def fourier (array):
    def rfunction (value):
        output = 0
        for i, a in enumerate(array):
            output += a * np.cos(2 * np.pi * i * value)
        return output
    return rfunction
```

```python
harmonics = np.array([140, 131, 120, 105, 86, 67, 48, 30, 14, 0])
four = fourier(harmonics)
x = np.linspace(-1.5, 1.5, 1000)
plt.plot(x, four(x))
```

```python
n = np.array([i for i in range(3, 13)])
t = np.array([0.68, 0.97, 1.21, 1.49, 1.60, 1.84, 2.14, 2.30, 2.54, 2.99])

f, tup = lsf(n, t)
print(tup)

plt.figure(figsize=(7, 4.9), dpi=300)

x = np.linspace(-1, 13, 3)
plt.scatter(n, t, marker='+')
plt.plot(x, f(x))

plt.grid(True)
plt.title("Зависимость $T(n)$")
plt.xlabel("$n$", loc="right")
plt.ylabel("$T$, сек", loc="top")

plt.xlim(-1, 13)
plt.ylim(-0.5, 3.5)

plt.savefig('plot1.png')
```
