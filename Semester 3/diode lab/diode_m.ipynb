#### Термоэлектронный диод

**Цель работы:** Практическое изучение термоэлектронной эмиссии и процессов токопрохождения в вакууме, 
изготовление вакуумного диода и исследование его вольтамперных характеристик

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['mathtext.fontset'] = 'dejavuserif' 
```

Прогрев катода

Погрешность $I_{нак} = 0.1 \text{А}$

Погрешность $U_{нак} = 0.1 \text{В}$

Погрешность $U_{анода} = 0.5 \text{В}$

Давление в камере перед прогревом: $5.3 \cdot 10^{-5}$

Давление в камере после прогрева: $ 7 \cdot 10^{-5}$

```python
cath = pd.DataFrame({
    "Iнак, А" : np.arange(0.1, 3.2, 0.1), 
    "Uнак, мВ" : np.array([22, 44.6, 69.2, 100.8, 143.1, 202, 278, 378, 483, 602, 693, 835, 1018, 1141, 1283, 1492, 1586, 2170, 2320, 2760, 2820, 3180, 3490, 3710, 4000, 4320, 4790, 5160, 5370, 5750, 6170])
})

plt.plot(cath.iloc[:, 0], cath.iloc[:, 1] / 10**(-3), color='magenta')
plt.xlabel(r"$I_{нак}$, А")
plt.ylabel(r"$U_{нак}$, В")
plt.title("Характеристика катода при прогреве")
plt.grid()

plt.plot()
```

```python
cath.iloc[:, 1]
```

```python
x = cath.iloc[:, 0]
y = cath.iloc[:, 1]

[a, b], res1 = curve_fit(lambda x1,a,b: a*np.exp(b*x1),  x,  y)
 
y1 = a * np.exp(b * x)

print(a, b)
 
plt.plot(x, y, 'b', label=r"$U_{нак} = 326 e^{0.98 I_{нак}}$")
plt.legend()
plt.plot(x, y1, 'r')
plt.show()
```
